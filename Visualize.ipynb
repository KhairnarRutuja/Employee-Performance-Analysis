```python
continuous_col = data[['Age', 'DistanceFromHome', 'EmpHourlyRate', 'EmpLastSalaryHikePercent', 'TotalWorkExperienceInYears', 'ExperienceYearsAtThisCompany', 'ExperienceYearsInCurrentRole', 'YearsSinceLastPromotion', 'YearsWithCurrManager']]

plt.figure(figsize=(16, 20), facecolor="white")
plotnumber = 1

for column in continuous_col:
    if plotnumber <= len(continuous_col):
        plt.subplot(5, 3, plotnumber)
        sns.set_style("darkgrid", {"grid.color": ".7", "grid.linestyle": ":"})
        sns.histplot(x=continuous_col[column], hue=data["PerformanceRating"], data=data, multiple="stack", kde=True, color='blue')
        plt.xlabel(column, fontsize=15)
        plt.ylabel("PerformanceRating", fontsize=15)
    plotnumber += 1

plt.tight_layout()
plt.show()
```

- Lower performance ratings at level 4 are observed in the range between 20 to 25 for salary hike percentage.

#### 5. RELATIONSHIP BETWEEN TOTAL WORKED EXPERIENCE IN YEARS  AND PERFORMANCE RATING
